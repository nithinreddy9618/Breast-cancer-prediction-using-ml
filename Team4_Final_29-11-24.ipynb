```python
X=cancer.iloc[:,2:32].values
```

```python
y=cancer.iloc[:,1].values
```

```python
from sklearn .model_selection  import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
from sklearn .preprocessing import StandardScaler
```

```python

#standardizes data by removing the mean and scaling to unit variance,
#improving machine learning algorithm performance.
st = StandardScaler()
```

```python
# to ensure consistent for both training and test data,preventing data leakage
#and improving model generalization

X_train = st.fit_transform(X_train)
X_test = st.transform(X_test)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
svc_model = SVC()
```

```python
from sklearn.metrics import accuracy_score
if isinstance(y_train, pd.Series):
    if y_train.dtype == 'Int64':
        y_train = y_train.astype(int)
    else:
        y_train = y_train.astype('category').cat.codes
elif y_train.dtype.kind not in 'iuf':
    from sklearn.preprocessing import LabelEncoder
    encoder = LabelEncoder()
    y_train = encoder.fit_transform(y_train)
else:
    y_train = y_train.astype(int)

svc_model.fit(X_train, y_train)
```

```python
y_pred = svc_model.predict(X_test)
```

```python
print(accuracy_score(pd.Series(y_test).astype('category').cat.codes, y_pred))
```
